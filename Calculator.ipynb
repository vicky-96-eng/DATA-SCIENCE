4**. Project**
**Option 1: Console-Based Smart Calculator**
*4.1 Problem Statement*
Create a menu-driven calculator that performs multiple mathematical operations using Python and
NumPy.
*4.2 Functional Requirements*
The calculator should: 1. Display a menu with operations 2. Take user choice as input 3. Ask for required
numbers 4. Perform calculation 5. Display result 6. Allow the user to continue or exit
*4.3 Mandatory Operations*
Addition
Subtraction
Multiplication
Division (handle division by zero)
Power calculation
Modulus operation



```python
import numpy as np

#menu
print("Calculator")
print("1. Addition")
print("2. Subtraction")
print("3. Multiplication")
print("4. Division")
print("5. Modulo")
print("6. Power operation")
print("7. Exit")
choice =0
while choice !=7: #if choice 7 then exit
    choice = int(input("\nEnter your choice: ")) #choice selection

    if choice == 1: #addition
        h = int(input("Enter how many numbers you want to add: "))
        l1 = []
        for i in range(h): #getting input & stored in list
            num = int(input("Enter number : "))
            l1.append(num)
        arr = np.array(l1) # create array using list
        ans = np.sum(arr)  # using sum of function
        print("Addition answer is:", ans)

    elif choice == 2:#subtraction
        a = int(input("Enter the first number: "))
        b = int(input("Enter the second number: "))
        ans = a - b #subtrating first to second
        print("Subtraction answer is:", ans)

    elif choice == 3: #multiply
        ans = 1
        print("Enter numbers to multiply (enter 0 to stop):")
        a=1
        while a!=0: #enter 0 to cut the loop
            a = int(input("Enter number: "))
            if a!=0:
                ans = ans * a # non-zero value getting multiplied
        print("Multiplication answer is:", ans)

    elif choice == 4:#division
        a = int(input("Enter numerator: "))
        b = int(input("Enter denominator: "))
        while b == 0: # checking numerator for non zero
            print("Division by zero Error. Enter a valid number.")
            b = int(input("Enter denominator: "))
        ans = a / b
        print("Division answer is:", ans)

    elif choice == 5: #modulo
        a = int(input("Enter first number: "))
        b = int(input("Enter second number: "))
        while b == 0: # checking numerator for non zero
            b = int(input("Division by zero. Enter valid number: "))
        ans = a % b
        print("Modulo answer is:", ans)

    elif choice == 6: #power
        a = int(input("Enter base value: "))
        b = int(input("Enter power value: "))
        ans = a ** b
        print("Power result:", ans)
    elif choice == 7: #Exit
        print("Exiting....")

    else:
        print("Invalid choice")
```
